```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Loan_default.csv')
df
```

```python
df = df.drop(columns= ['LoanID','LoanPurpose','HasCoSigner'])
```

```python
df
```

```python
df['Education'] = df['Education'].astype('category')
df['Education'] = df['Education'].cat.codes
df
```

```python
df['EmploymentType'] = df['EmploymentType'].astype('category')
df['EmploymentType'] = df['EmploymentType'].cat.codes

df['MaritalStatus'] = df['MaritalStatus'].astype('category')
df['MaritalStatus'] = df['MaritalStatus'].cat.codes

df['HasMortgage'] = df['HasMortgage'].astype('category')
df['HasMortgage'] = df['HasMortgage'].cat.codes

df['HasDependents'] = df['HasDependents'].astype('category')
df['HasDependents'] = df['HasDependents'].cat.codes

df
```

```python
df.isnull().sum()
```

```python
X = df.drop(columns='Default')
```

```python
X
```

```python
Y = df['Default']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test =  train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train) 
#fit transform stores and uses the mean and SD for fitting of the model
```

```python
X_test_scaled = scaler.transform(X_test) 
#transform does not stores the mean and SD because we do not need the assumptions from the test datset
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(random_state=0).fit(X_train_scaled, Y_train)
```

```python
log_reg.predict(X_train_scaled)
```

```python
log_reg.score(X_train_scaled, Y_train)
```

```python
log_reg.score(X_test_scaled, Y_test)
```

```python
log_reg = LogisticRegression(random_state=0,
                            fit_intercept=True
                            ).fit(X_train_scaled, Y_train)
```

```python
log_reg.score(X_train_scaled, Y_train)
```

```python
log_reg.score(X_test_scaled, Y_test)
```

```python
Y_pred = log_reg.predict(X_test_scaled)
```

```python
Y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

#Confusion Matrix

CM = confusion_matrix(Y_test,Y_pred)
print('Confusion Matrix : ', CM)
```

```python
#Accuracy score
Accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy Score : ', Accuracy)
```

```python
#Precision Score
Precision = precision_score(Y_test, Y_pred)
print('Precision Score : ', Precision)
```

```python
#Recall Score

Recall = recall_score(Y_test,Y_pred)
print('Recall Score : ', Recall)
```

```python
Y_prob = log_reg.predict_proba(X_test_scaled)[:,1]

Y_prob
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

fpr, tpr, thresholds = roc_curve(Y_test,Y_prob)
```

```python
plt.plot(fpr, tpr, label="ROC Curve")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.legend()
plt.show()
```

```python
AUC = roc_auc_score(Y_test, Y_prob)

print('AUC Score : ',AUC)
```
